##### Visualize the similarity of different model results

```python
def jaccard_similarity(set_a, set_b):
    intersection = len(set_a.intersection(set_b))
    union = len(set_a.union(set_b))
    return intersection / union
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
def plot_topics_heatmap(model_a, model_b):

    model_a_topics = globals().get(f"{model_a}_topics")
    model_b_topics = globals().get(f"{model_b}_topics")
    
    labels_a = globals().get(f"{model_a}_topic_name")
    labels_b = globals().get(f"{model_b}_topic_name")
    
    # Initialize an empty matrix
    similarity_matrix = [[0 for _ in range(len(model_b_topics))] for _ in range(len(model_a_topics))]
    
    # Calculate Jaccard similarity for each pair of topics
    for i, topic_a in enumerate(model_a_topics):
        for j, topic_b in enumerate(model_b_topics):
            similarity_matrix[i][j] = jaccard_similarity(topic_a, topic_b)
    
    # Convert the similarity matrix to a DataFrame for easier plotting
    similarity_df = pd.DataFrame(similarity_matrix)
    
    # Create a heatmap using seaborn
    plt.figure(figsize=(10, 8))
    sns.heatmap(similarity_df, annot=False, cmap="YlGnBu", fmt=".2f",xticklabels=labels_b, yticklabels=labels_a)
    
    # Add labels and a title
    plt.xlabel(f"{model_b} Topics")
    plt.ylabel(f"{model_a} Topics")
    plt.title(f"Jaccard Similarity between {model_a} and {model_b} Topics")
    plt.show()
```

##### Plot topics similarity for LDA model and Bert model

```python
plot_topics_heatmap("LDA", "BERT")
```

#### Plot topics similarity for Bert model and NMF model

```python
plot_topics_heatmap("BERT", "NMF")
```
